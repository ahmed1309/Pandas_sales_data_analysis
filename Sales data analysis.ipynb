```python
yearly_data = pd.read_csv("yearly_data.csv")
yearly_data.head()
```

# Augment the data with additionat coloumns

## add month coloumn

```python
yearly_data['Month'] = yearly_data['Order Date'].str[0:2]
yearly_data.head()
```

# Clean Data

## Drop rows with NaN

```python
yearly_data = yearly_data.dropna()
```

## excluding rows where Or is present

```python
yearly_data = yearly_data[yearly_data['Order Date'].str[0:2] != 'Or']
yearly_data.head()
```

## Converting month type to int

```python
yearly_data['Quantity Ordered'] = pd.to_numeric(yearly_data['Quantity Ordered'])
yearly_data['Price Each'] = pd.to_numeric(yearly_data['Price Each'])
```

## Converting month type to int

```python
yearly_data.head()
```

```python
yearly_data['Month'] = yearly_data['Month'].astype(int)
yearly_data.head()
```

# Per month Earning Calculation

## Adding a coloumn of sales 

```python
yearly_data.insert(4, 'Sales', 0)
```

### Sales = (Quantity Ordered * Price Each)

```python
yearly_data['Sales'] = yearly_data['Quantity Ordered']*yearly_data['Price Each']
yearly_data.head()
```

```python
results = yearly_data.groupby('Month').sum()
results.head(15)
```

```python
import matplotlib.pyplot as plt
months = range(1,13) 
plt.xticks(months)
plt.xlabel('months')
plt.ylabel('sales in USD(10^6)')
plt.bar(months,results['Sales'])
```

# Region dependent analysis

## Adding a coloumn of City 

```python
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]
yearly_data['City'] = yearly_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})" )
yearly_data.head()
```

```python
cityresults = yearly_data.groupby('City').sum()
cityresults.head(15)
```

```python
cities = [city for city, df in yearly_data.groupby('City')] 
#yearly_data['City'].unique()  #.....this also works, but it changes the order of cities which swaps the values in plot
plt.bar(cities,cityresults['Sales'])
plt.xticks(cities, rotation = 'vertical',size=8)
plt.xlabel('Cities')
plt.ylabel('Sales in USD X(10^6)')
```

# Maximum customer hour to optimize the advertisement time

```python
yearly_data['Order Date'] = pd.to_datetime(yearly_data['Order Date'])
yearly_data.head()
```

```python
yearly_data['Hour'] = yearly_data['Order Date'].dt.hour
yearly_data['Minute'] = yearly_data['Order Date'].dt.minute
yearly_data.head()
```

```python
# in this case instead of sum() we will count(). because we have to count the number of orders.
hourly_results = yearly_data.groupby('Hour').count()
hourly_results.head(30)
```

```python
hours = [hour for hour, df in yearly_data.groupby('Hour')]
yearly_data.groupby('Hour').count()
plt.plot(hours , yearly_data.groupby('Hour').count()['Order Date'])
plt.xticks(hours)
plt.grid()
#plt.plot(hours , hourly_results['Order Date'])
```

# Most Sold Product

```python
product_grouping = yearly_data.groupby('Product').sum()
product_grouping.head()
```

```python
prices = yearly_data.groupby('Product').mean()['Price Each']
```

```python
products = [product for product, df in yearly_data.groupby('Product') ]
fig , ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.bar(products, product_grouping['Quantity Ordered'])
ax2.plot(products, prices, 'g-')

ax1.set_xticklabels(products, rotation = 'vertical',size=8)
ax1.set_xlabel('Products')
ax1.set_ylabel('Quantity Sold', color = 'b')
ax2.set_ylabel('Price($)', color = 'g')
plt.show()
```

# Products which were sold together

```python
grouping = yearly_data[yearly_data['Order ID'].duplicated(keep=False)]
grouping['Grouped'] = grouping.groupby('Order ID')['Product'].transform(lambda x: ','.join(x)) 
grouping = grouping[['Order ID','Grouped']].drop_duplicates()
grouping.head(20)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()
for row in grouping['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,4)))
for key, value in count.most_common(10):
    print("Products", key, "were sold together", value, "times")
```
